```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
```

```python
d = np.loadtxt("atermonly2.txt", usecols=(0,1))
plt.plot(d[:,0], d[:,1])
plt.xlabel("Wavelength (nm)")
plt.ylabel("Absorbance")
plt.xlim(140, 200)
plt.savefig("aTermOnly2.png", dpi=200, format="png")
```
